```
import numpy as np
from numpy import genfromtxt

err_a = genfromtxt('/content/S-DCNet-master/error_data_SHA.out', delimiter=',')
err_b = genfromtxt('/content/S-DCNet-master/error_data_SHB.out', delimiter=',')
```

```
import matplotlib.pyplot as plt

def plot(arr, title = None, abs_val = False):
  error_vals = arr[1]
  if abs_val is True:
    error_vals = abs(error_vals)
  plt.scatter(arr[0], error_vals, alpha = 0.4, s = abs(arr[1]))
  plt.title(title)
  plt.xlabel('Ground Truth')
  plt.ylabel('Absolute Error')
  plt.show()

plot(err_a, "Error for SHA - ABS", abs_val = True)
plot(err_a, "Error for SHA", abs_val = False)

plot(err_b, "Error for SHB - ABS", abs_val = True)
plot(err_b, "Error for SHB", abs_val = False)
```
